# Antibiotic Resistance Data Cleaning

```python
import pandas as pd 
pd.set_option("display.max_columns", 30)
```

```python
df = pd.read_excel('../data/Questionnaire_Antibiotic_Resistance.xlsx')  
```

```python
df.head() 
```

```python
# remove () with bangla text 
df.columns = df.columns.str.replace(r"\([^()]*\)","")
```

```python
# remove : 
df.columns = df.columns.str.replace(':','')
```

```python
# remove ? at the end 
df.columns = df.columns.str.replace('?','')
```

```python
df.columns = df.columns.str.strip() 
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```
